Ahora vamos a filtrar los casos confirmados. Esto es, los caso que tienen en la columna CLASIFICACION_FINAL un valor de 1, 2 o 3.

Por último los casos sospechosos.

Notamos que en la columna 'CLASIFICACION_FINAL' hay resultados clasificados como "Inválido por laboratorio", "rechazado por laboratorio" y "sospechoso" par alos cuales el resultado de laboratorio no es adecuado. Esto es son casos sospechosos sin posibilidad de resultado.

```python
sospechosos_sin_posibilidad = data[(data['RESULTADO_LAB'] == 4) & (data['CLASIFICACION_FINAL'].isin([6,5,4]))]
```

Para registrar los decesos la base de datos no nos proporciona la información deseada. Pero de manera indirecta podemos obtender este dato viendo que la  base de datos tiene una columna "fecha de defunción". Primero exploramos esta columna.

```python
data['FECHA_DEF'].value_counts()
```

Con base a este análisis podemos deducir que al capturar los casos, se captura 9999-99-99 cuando la persona no ha fallecido.

```python
defunciones = data[data['FECHA_DEF'] != "9999-99-99"]
defunciones
```

```python
grupos = positivos.groupby("FECHA_SINTOMAS")
confirmados_por_fecha = grupos.size().to_frame("confirmados").reset_index()
```

Para poder graficar necesitamos dos columnas, por lo que "añadimos" una columna de índices. Antes de poder graficar necesitamos ver en que tipo de datos están guardados los datos.

```python
confirmados_por_fecha
```

```python
confirmados_por_fecha.dtypes
```

Entonces buscamos convertir los datos de la columna "Fecha de síntomas" de tipo object a datos de tipo fecha. Y agregarlos como una nueva columna a nuestro Dataframe.

```python
confirmados_por_fecha["fecha_sintomas"] = pd.to_datetime(confirmados_por_fecha["FECHA_SINTOMAS"])
```

```python
confirmados_por_fecha
```

Checamos que los tipos de datos sean correctos

```python
confirmados_por_fecha.dtypes
```

Repetimos este proceso para nuestros DF de defunciones y sospechosos.

```python
grupo_def = defunciones.groupby("FECHA_SINTOMAS")
defunciones_por_fecha = grupo_def.size().to_frame("defunciones").reset_index()
defunciones_por_fecha["fecha_sintomas"] = pd.to_datetime(defunciones_por_fecha["FECHA_SINTOMAS"])
defunciones_por_fecha
```

```python
grupo_sos = sospechosos.groupby("FECHA_SINTOMAS")
sospechosos_por_fecha = grupo_sos.size().to_frame("sospechosos").reset_index()
sospechosos_por_fecha["fecha_sintomas"] = pd.to_datetime(sospechosos_por_fecha["FECHA_SINTOMAS"])
sospechosos_por_fecha
```

# Gráficas

```python
plt.plot(confirmados_por_fecha["fecha_sintomas"], confirmados_por_fecha["confirmados"])
plt.title("Casos confirmados por fecha")
plt.xlabel("Fecha")
plt.ylabel("Casos confirmados")
plt.show()
```

```python
plt.plot(defunciones_por_fecha["fecha_sintomas"], defunciones_por_fecha["defunciones"])
plt.title("defunciones por fecha")
plt.xlabel("Fecha")
plt.ylabel("defunciones")
plt.show()
```

```python
plt.plot(sospechosos_por_fecha["fecha_sintomas"], sospechosos_por_fecha["sospechosos"])
plt.title("Sospechosos por fecha")
plt.xlabel("Fecha")
plt.ylabel("Casos sospechosos")
plt.show()
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(3,1,1)
ax1.plot(confirmados_por_fecha["fecha_sintomas"], confirmados_por_fecha["confirmados"])
ax2 = fig.add_subplot(3,1,2)
ax2.plot(defunciones_por_fecha["fecha_sintomas"], defunciones_por_fecha["defunciones"])
ax3 = fig.add_subplot(3,1,3)
ax3.plot(sospechosos_por_fecha["fecha_sintomas"], sospechosos_por_fecha["sospechosos"])
plt.show()
```

```python
plt.plot(confirmados_por_fecha["fecha_sintomas"], confirmados_por_fecha["confirmados"], label="Confirmados")
plt.plot(defunciones_por_fecha["fecha_sintomas"], defunciones_por_fecha["defunciones"], label="Defunciones")
plt.plot(sospechosos_por_fecha["fecha_sintomas"], sospechosos_por_fecha["sospechosos"], label="Sospechosos")
plt.title("Datos Covid-19 Nacionales")
plt.xlabel("Fecha")
plt.ylabel("Casos")
fig = plt.gcf()
fig.set_size_inches(16,10)
plt.legend()
plt.savefig("nac1.jpg", facecolor='w', edgecolor='w')
plt.show()
```

```python
plt.plot(confirmados_por_fecha["fecha_sintomas"], confirmados_por_fecha["confirmados"], "o",label="Confirmados")
plt.plot(defunciones_por_fecha["fecha_sintomas"], defunciones_por_fecha["defunciones"], "o",label="Defunciones")
plt.plot(sospechosos_por_fecha["fecha_sintomas"], sospechosos_por_fecha["sospechosos"], "o",label="Sospechosos")
plt.title("Datos Covid-19")
plt.xlabel("Fecha")
plt.ylabel("Casos")
fig = plt.gcf()
fig.set_size_inches(16,10)
plt.legend()
plt.savefig("nac2.jpg", facecolor='w', edgecolor='w')
plt.show()
```

```python
B = np.copy(sospechosos_por_fecha["sospechosos"])
B.fill(0000)
A = np.copy(sospechosos_por_fecha["sospechosos"])
A.fill(200000)
```

```python
plt.plot(confirmados_por_fecha["fecha_sintomas"], confirmados_por_fecha["confirmados"].cumsum(), label="Confirmados")
plt.plot(defunciones_por_fecha["fecha_sintomas"], defunciones_por_fecha["defunciones"].cumsum(), label="Defunciones")
plt.plot(sospechosos_por_fecha["fecha_sintomas"], sospechosos_por_fecha["sospechosos"].cumsum(), label="Sospechosos")
plt.plot(sospechosos_por_fecha["fecha_sintomas"], B, "r--", label  = "1.75 millones")
plt.plot(sospechosos_por_fecha["fecha_sintomas"], A, "b--", label  = "200 mil")

plt.title("Datos Covid-19 nacionales")
plt.xlabel("Fecha")
plt.ylabel("Casos acumulados")
fig = plt.gcf()
fig.set_size_inches(16,10)
plt.legend()
plt.savefig("naccum.jpg", facecolor='w', edgecolor='w')
plt.show()
```

```python
defunciones_estado =data_estado[data_estado["FECHA_DEF"] != "9999-99-99"]
defunciones_estado
```

```python
conf_fechas_estado = grupo_estado.size().to_frame("confirmados").reset_index()
conf_fechas_estado['fecha_sintomas'] = pd.to_datetime(conf_fechas_estado["FECHA_SINTOMAS"])
conf_fechas_estado
```

```python
sosp_fechas_estado = sospechosos_estado.groupby("FECHA_SINTOMAS").size().to_frame("sospechosos").reset_index()
sosp_fechas_estado['fecha_sintomas'] = pd.to_datetime(sosp_fechas_estado["FECHA_SINTOMAS"])
sosp_fechas_estado
```

```python
def_fechas_estado = defunciones_estado.groupby("FECHA_SINTOMAS").size().to_frame("defunciones").reset_index()
def_fechas_estado['fecha_sintomas'] = pd.to_datetime(def_fechas_estado["FECHA_SINTOMAS"])
def_fechas_estado
```

```python
plt.plot(conf_fechas_estado["fecha_sintomas"], conf_fechas_estado["confirmados"], label = "Confirmados")
plt.plot(def_fechas_estado["fecha_sintomas"], def_fechas_estado["defunciones"], label = "Defunciones")
plt.plot(sosp_fechas_estado["fecha_sintomas"], sosp_fechas_estado["sospechosos"], label = "Sospechosos")
plt.title("Datos Covid-19, estado de Durango")
plt.xlabel("Fecha de sintomas")
plt.ylabel("Casos")
fig = plt.gcf()
fig.set_size_inches(14,10)
plt.grid(True)
plt.legend()
plt.savefig("estado.jpg", facecolor='w', edgecolor='w')
plt.show()
```

```python
plt.plot(conf_fechas_estado["fecha_sintomas"], conf_fechas_estado["confirmados"].cumsum(), label = "Confirmados")
plt.plot(def_fechas_estado["fecha_sintomas"], def_fechas_estado["defunciones"].cumsum(), label = "Defunciones")
plt.plot(sosp_fechas_estado["fecha_sintomas"], sosp_fechas_estado["sospechosos"].cumsum(), label = "Sospechosos")
plt.title("Datos Covid-19 Durango.")
plt.xlabel("Fecha")
plt.ylabel("Acumulado de casos")
fig = plt.gcf()
fig.set_size_inches(16,10)
plt.grid(True)
plt.legend()
plt.savefig("estadocum.jpg", facecolor='w', edgecolor='w')
plt.show()
```
